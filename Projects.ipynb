Missionaries & Cannibals

```python
from IPython.display import clear_output

boat_side='r'
missionaries_on_right=3
cannibals_on_right=3
missionaries_on_left=0
cannibals_on_left=0

print(missionaries_on_left* '\U0001f482',cannibals_on_left*'\U0001f479','|\U0001f30a\U0001f30a\U0001f30a\U0001f30a\U0001f6A2|',missionaries_on_right* '\U0001f482',cannibals_on_right*'\U0001f479')


while True:
    missionaries=int(input("missionaries on the boat : "))
    cannibals=int(input("cannibals on the boat : "))
    clear_output()


    
    if boat_side=='r':
        print("boat is on right side")
        clear_output()


        if missionaries+cannibals==1 or missionaries+cannibals==2:
            if missionaries>missionaries_on_right or cannibals>cannibals_on_right:
                print("Invalid Move 2")
                continue
            else:
                missionaries_on_right=missionaries_on_right-missionaries
                cannibals_on_right=cannibals_on_right-cannibals
                missionaries_on_left=missionaries_on_left+missionaries
                cannibals_on_left=cannibals_on_left+cannibals

                print(missionaries_on_left* '\U0001f482',cannibals_on_left*'\U0001f479','|\U0001f6A2\U0001f30a\U0001f30a\U0001f30a\U0001f30a|',missionaries_on_right* '\U0001f482',cannibals_on_right*'\U0001f479')
                if missionaries_on_right<cannibals_on_right and missionaries_on_right>0:
                  print('You lose')
                  break
                elif missionaries_on_left<cannibals_on_left and missionaries_on_left>0:
                  print('You lose')
                  break
                elif missionaries_on_left==3 and cannibals_on_left==3:
                  print('You Win')
                  break
                boat_side='l'
        else:
            print("Invalid Move")
            continue

    
    elif boat_side=='l':
        print("boat is on left side")
        clear_output()


        if missionaries+cannibals==1 or missionaries+cannibals==2:
            if missionaries>missionaries_on_left or cannibals>cannibals_on_left:
                print("Invalid Move 2")
                continue
            else:
                missionaries_on_left-=missionaries
                cannibals_on_left-=cannibals
                missionaries_on_right+=missionaries
                cannibals_on_right+=cannibals

                print(missionaries_on_left* '\U0001f482',cannibals_on_left*'\U0001f479','|\U0001f30a\U0001f30a\U0001f30a\U0001f30a\U0001f6A2|',missionaries_on_right* '\U0001f482',cannibals_on_right*'\U0001f479')
                boat_side='r'
                if missionaries_on_right<cannibals_on_right and missionaries_on_right>0:
                  print('You lose')
                  break
                elif missionaries_on_left<cannibals_on_left and missionaries_on_left>0:
                  print('You lose')
                  break
        else:
            print('invalid move')
            continue
                
print('Game Over')
```

Frog Leap

```python
position=['G', 'G', 'G', '-', 'B', 'B', 'B']
print(' [ 0 ,  1 ,  2 ,  3 ,  4 ,  5 ,  6 ]\n',position)
while True:
    inp1=input("Press q to quit else \n Enter position of piece:")
    if inp1=='q':
        print('You Lose')
        break
    else:
        inp=int(inp1)
    if inp<0 and inp>6:
        print('!!!!!!!!Invalid move!!!!!!!!')
    if position[inp]=='-':
        print('!!!!!!!!Invalid move!!!!!!!!')
    pos2=0
    if position[inp]=='G':
        if inp+1<=6 and position[inp+1]=='-':
            pos2=inp+1
        elif inp+2<=6 and position[inp+2]=='-' and position[inp+1]=='B' :
            pos2=inp+2
        else:
            print('!!!!!!!!Invalid move!!!!!!!!')
    if position[inp]=='B':
        if inp-1>=0 and position[inp-1]=='-':
            pos2=inp-1
        elif inp-2>=0 and position[inp-2]=='-' and position[inp-1]=='G':
            pos2=inp-2
        else:
            print('!!!!!!!!Invalid move!!!!!!!!')
    
    position[inp],position[pos2]=position[pos2],position[inp]
    
    print(' [ 0 ,  1 ,  2 ,  3 ,  4 ,  5 ,  6 ]\n',position)
    
    if position== ['B', 'B', 'B','-','G', 'G','G']:
        print('You Win')
        break
```

Contact App

```python
from IPython.display import clear_output
def show_contacts(contacts):
    for i in range(len(contacts)):
        print(contacts[i]['name'])
def add_contact(contacts):
        l1=['name','number','email']
        d1={i:input() for i in l1}
        contacts.append(d1)
def delete_contact(contacts,name):
    for i in range(len(contacts)):
        if contacts[i]['name']==name:
            del contacts[i]
def update_contact(contacts, name):
    for k in range(len(contacts)):
        if contacts[k]['name']==name:
            contacts[k]['name']=input('NewName : ')
            contacts[k]['number']=input('Newnumber : ')
            contacts[k]['email']=input('Newemail : ')
def open_contact(contacts, name):
    for i in range(len(contacts)):
        if contacts[i]['name']==name:
            print(contacts[i])
            input1=input('press u to update & d delete contact. ')
            if input1=='u':
                update_contact(contacts, name)
            elif input1=='d':
                delete_contact(contacts,name)
            else:
                pass
        else :
            print( "No match found!")      

while True:
    show_contacts(contacts)
    action = input("Press 'a' - add contact, 'o' - open contact,'s'- search contact, 'q' - quit")
    clear_output()
    if action == 'a':
        add_contact(contacts)
    elif action == 'o':
        name = input("Enter name of the contact you want to open. ")        
        open_contact(contacts, name)
    elif action=='s':
        name1=input("Enter name of the contact you want to search. " )
        for i in range(len(contacts)):
            if contacts[i]['name']==name1:
                print("Contact found!")
                print(contacts[i])
    elif action == 'q':
        break
    else:
        print("Incorrect choise!")
```
